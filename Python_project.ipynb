###  4. You need to analyze the monthly profit from sales and determine which month had the highest profit.

```python
sm["Order Date"] = pd.to_datetime(sm["Order Date"])
sm["Month"] = sm["Order Date"].dt.month_name()

monthly_profit = sm.groupby("Month")["Profit"].sum()
monthly_profit = monthly_profit.reindex([
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
])

print("Monthly Profit:\n", monthly_profit)

best_month = monthly_profit.idxmax()
best_value = monthly_profit.max()
print(f"\nHighest Profit: {best_month} (${best_value:,.2f})")

plt.figure(figsize=(10,5))
monthly_profit.plot(kind='bar')
plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###  6. Analyze the sales and profit by customer segment

### 7. Analyze the sales to profit ratio

```python
sm["Sales_Profit_Ratio"] = sm["Sales"] / sm["Profit"].replace(0, pd.NA)

ratio_data = sm.groupby("Sub-Category")["Sales_Profit_Ratio"].mean().sort_values()

print("\nSales-to-Profit Ratio by Sub-Category:")
print(ratio_data)

plt.figure(figsize=(10,6))
ratio_data.plot(kind="line")
plt.title("Average Sales-to-Profit Ratio by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Sales / Profit Ratio")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
